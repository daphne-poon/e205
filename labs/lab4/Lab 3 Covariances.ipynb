```python
import pandas as pd
import numpy as np
import matplotlib as plt
from scipy.stats import norm
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv("./data/2020_2_26__16_59_7_filtered.csv")
a=len(df.columns)-1 
df=df.iloc[:,:a]
data = pd.DataFrame(df.values, columns = ["X", "Y", "Z", "Timestamp", "Latitude", "Longitude",
                                         "Yaw", "Pitch", "Roll", "AccelX", "AccelY", "AccelZ"])
```

```python
# For X acceleration

(mu, sigma) = norm.fit(data["AccelX"][4:44])
plt.hist(data["AccelX"][4:44],bins=20, density=True)
# sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, norm.pdf(x, mu, sigma))
plt.xlabel("Measured X Acceleration (m/s^2)")
plt.ylabel("Probability / Normalized Frequency")
plt.ylabel("Probability / Normalized Frequency")
mu, sigma
```

```python
# For Y acceleration

(mu, sigma) = norm.fit(data["AccelY"][4:44])
plt.hist(data["AccelX"][4:44],bins=20, density=True)
# sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, norm.pdf(x, mu, sigma))
mu, sigma
```

```python
# For Yaw

def wrap_to_pi(angle):
    """Wrap angle data in degrees to [-pi, pi]

    Parameters:
    angle (float)   -- unwrapped angle

    Returns:
    angle (float)   -- wrapped angle
    """
    # convert to rad
    angle = angle*(math.pi)/180
    # wrap
    while angle >= math.pi:
        angle -= 2*math.pi

    while angle <= -math.pi:
        angle += 2*math.pi
    return angle

#process
nums = data["Yaw"][4:44]
processed = list(map(wrap_to_pi, nums))

(mu, sigma) = norm.fit(processed)
plt.hist(processed, bins=20, density=True)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, norm.pdf(x, mu, sigma))
mu, sigma
```
